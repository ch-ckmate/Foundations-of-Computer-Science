# For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year

```python
loans['planned_expiration_year']=pd.to_datetime(loans['planned_expiration_time']).dt.year
loans['disburse_year']=pd.to_datetime(loans['disburse_time']).dt.year
df00=loans[['loan_amount','disburse_year','planned_expiration_year','disburse_time', 'planned_expiration_time']]
foreachyear = df00.groupby('disburse_year')['loan_amount'].sum()#sum oppure count
foreachyear
```

```python
df00['giorno_dellanno']=loans['disburse_time'].dt.dayofyear
df00['giorno_finale']=loans['planned_expiration_time'].dt.dayofyear
df00['stesso_anno']=df00['planned_expiration_year']-df00['disburse_year']
#Considero solo quelli > di zero dato che devono avere anni diversi disburse e planned
df01 = df00[df00['stesso_anno']>0]
df01
```

```python
df02=df01.dropna(subset=['stesso_anno'])
df02.sort_values(by=['disburse_year'])
df03=df02.drop(columns=['disburse_time','planned_expiration_time'])
```

```python
df03['2011']=0
df03['2012']=0
df03['2013']=0
df03['2014']=0
df03['2015']=0
df03['2016']=0
df03['2017']=0
df03['2018']=0
```

```python
#Differsico per gli anni bisestili
df03.loc[df03['disburse_year'] == 2011, 'giorno_di_inizio'] = 365 - df03['giorno_dellanno']
df03.loc[df03['disburse_year'] == 2012, 'giorno_di_inizio'] = 366 - df03['giorno_dellanno']
df03.loc[df03['disburse_year'] == 2013, 'giorno_di_inizio'] = 365 - df03['giorno_dellanno']
df03.loc[df03['disburse_year'] == 2014, 'giorno_di_inizio'] = 365 - df03['giorno_dellanno']
df03.loc[df03['disburse_year'] == 2015, 'giorno_di_inizio'] = 365 - df03['giorno_dellanno']
df03.loc[df03['disburse_year'] == 2016, 'giorno_di_inizio'] = 366 - df03['giorno_dellanno']
df03.loc[df03['disburse_year'] == 2017, 'giorno_di_inizio'] = 365 - df03['giorno_dellanno']
```

```python
df03.loc[100:200]
```

```python
def loan(x):
    if x['stesso_anno']== 1:
        if x['disburse_year'] == 2012 or x['disburse_year'] == 2016:            
            x[str(int(x['disburse_year']))]= (x['giorno_di_inizio']*x['loan_amount'])/((x['giorno_di_inizio'])+ x['giorno_finale']+ ((x['stesso_anno']-1)*366))
            x[str(int(x['planned_expiration_year']))]= x['loan_amount'] - x[str(int(x['disburse_year']))]
        else:
            x[str(int(x['disburse_year']))]= (x['giorno_di_inizio']*x['loan_amount'])/((x['giorno_di_inizio'])+ x['giorno_finale']+ ((x['stesso_anno']-1)*366))
            x[str(int(x['planned_expiration_year']))]= x['loan_amount'] - x[str(int(x['disburse_year']))]
    elif x['stesso_anno'] > 1:
        if x['disburse_year'] == 2012 or x['disburse_year'] == 2016:            
            x[str(int(x['disburse_year']))]= x['giorno_di_inizio']*x['loan_amount'] / (x['giorno_di_inizio'] + x['giorno_finale'] + ((x['stesso_anno']-1)*366))
            x[str(int(x['planned_expiration_year']))]=(x['giorno_finale']*x['loan_amount'])/((x['giorno_di_inizio'])+ x['giorno_finale']+ ((x['stesso_anno']-1)*366))
        else:
            x[str(int(x['disburse_year']))]= x['giorno_di_inizio']*x['loan_amount'] / (x['giorno_di_inizio'] + x['giorno_finale'] + ((x['stesso_anno']-1)*365))
            x[str(int(x['planned_expiration_year']))]=(x['giorno_finale']*x['loan_amount'])/((x['giorno_di_inizio'])+ x['giorno_finale']+ ((x['stesso_anno']-1)*365))
            for y in range(1,int(x['stesso_anno'])):
                if x['disburse_year']+y == 2012 or x['disburse_year']+y == 2016:
                    
                    x[str(int(x['disburse_year']+y))] = (366*x['loan_amount'])/((x['giorno_di_inizio'])+ x['giorno_finale']+ ((x['stesso_anno']-1)*366))
                else:
                    x[str(int(x['disburse_year']+y))] = (365*x['loan_amount'])/((x['giorno_di_inizio'])+ x['giorno_finale']+ ((x['stesso_anno']-1)*365))

    return x
```
